```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import *
spark = SparkSession.builder.appName('Demo2').getOrCreate()
```

Now, I'm going to read the datasets that I'm considering, i.e. "litecoin price" and "ethereum price".

```python
#header specifies each column and inferschema is used to automatically detect data types, where 'lp' represents dataset for litecoin, and 'ep' represents dataset for ethereum.
lp = spark.read.csv('Dataset_main/litecoin_price (1).csv',header=True,inferSchema=True)
ep = spark.read.csv('Dataset_main/ethereum_classic_price 1.csv',header=True,inferSchema=True)
```

Now, I will check how many null values my dataset consist.

```python
#Reading data before checking for null values.
lp=pd.read_csv('Dataset_main/litecoin_price (1).csv',index_col='Date')
#Checking null values.
null_counts = lp.isnull().sum()
null_counts[null_counts > 0].sort_values(ascending=False)
```

```python
lp.fillna(lp.mean())
```

The output shows that the volume column in litecoin dataset contains 244 null values.
Now, I will check the same for ethereum dataset.

```python
#Reading data before checking for null values.
ep=pd.read_csv('Dataset_main/ethereum_classic_price 1.csv',index_col='Date')
#Checking null values.
null_counts = ep.isnull().sum()
null_counts[null_counts > 0].sort_values(ascending=False)
```

The output shows that the volume column in ethereum dataset contains 244 null values and market cap consist 1 null value.

```python
#Merging both tables for better exploration. I read the dataset and carry out an outer join by using the common column 'Date'.
import pandas as pd
lp=pd.read_csv('Dataset_main/litecoin_price (1).csv',index_col='Date')
lp.head()
ep=pd.read_csv('Dataset_main/ethereum_classic_price 1.csv',index_col='Date')
ep.head()
Merged_Pandas=lp.merge(ep, on='Date', how='outer')
Merged_Pandas.head()
```

In the above merged table, column name followed by _x represents the attributes from litecoin dataset whereas column name followed by _y represents the attributes from ethereum dataset.

```python
#Visualising the merged data by plotting 'High' for litecoin on x axis and plotting 'High' for ethereum on y axis.
Merged_Pandas.plot(x='High_x', y="High_y")
```
