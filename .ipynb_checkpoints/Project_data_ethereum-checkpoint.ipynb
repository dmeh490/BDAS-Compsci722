```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import *
spark = SparkSession.builder.appName('Project_data_ethereum').getOrCreate()
```

```python
ep = spark.read.csv('Dataset_main/ethereum_classic_price 1.csv',header=True,inferSchema=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# logarithm doesn't allow value 0, so increase 0.0001 for each value
new_ep = ep.withColumn('High_',ep['High']+0.0001)
new_ep = new_ep.toPandas()

np.log(new_ep['High_']).plot.hist(bins=50)
plt.title('High Logarithm')
plt.ylabel('')

plt.show()
```

```python
new_ep = ep.withColumn('Open_',ep['Open']+0.0001)
new_ep = new_ep.toPandas()

np.log(new_ep['Open_']).plot.hist(bins=50)
plt.title('Open Logarithm')
plt.ylabel('')

plt.show()
```

```python
new_ep = ep.withColumn('Low_',ep['Low']+0.0001)
new_ep = new_ep.toPandas()

np.log(new_ep['Low_']).plot.hist(bins=50)
plt.title('Low Logarithm')
plt.ylabel('')

plt.show()
```

```python
new_ep = ep.withColumn('Close_',ep['Close']+0.0001)
new_ep = new_ep.toPandas()

np.log(new_ep['Close_']).plot.hist(bins=50)
plt.title('Close Logarithm')
plt.ylabel('')

plt.show()
```
