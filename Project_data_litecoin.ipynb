```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import *
spark = SparkSession.builder.appName('Project_data_litecoin').getOrCreate()
```

```python
lp = spark.read.csv('Dataset_main/litecoin_price (1).csv',header=True,inferSchema=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# logarithm doesn't allow value 0, so increase 0.0001 for each value
new_lp = lp.withColumn('High_',lp['High']+0.0001)
new_lp = new_lp.toPandas()

np.log(new_lp['High_']).plot.hist(bins=50)
plt.title('High Logarithm')
plt.ylabel('')

plt.show()
```

```python
new_lp = lp.withColumn('Low_',lp['Low']+0.0001)
new_lp = new_lp.toPandas()

np.log(new_lp['Low_']).plot.hist(bins=50)
plt.title('Low Logarithm')
plt.ylabel('')

plt.show()
```

```python
new_lp = lp.withColumn('Open_',lp['Open']+0.0001)
new_lp = new_lp.toPandas()

np.log(new_lp['Open_']).plot.hist(bins=50)
plt.title('Open Logarithm')
plt.ylabel('')

plt.show()
```
